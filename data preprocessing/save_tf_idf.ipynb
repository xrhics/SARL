```python
def get_df_map(data):
    DF_map={}
    for i in tqdm(range(1,len(data.loc[0]))):
        exist = (data.loc[:,str(i)] > 0) * 1
        DF_map[i]=sum(exist)
    return DF_map
```

```python
import pandas as pd
from tqdm import tqdm
import numpy as np
from math import log
lengths = [200, 300, 400, 500]
cities = ["Milano", "Amsterdam", "Barcelona", "Lisboa"]
min_count=5
for city in cities:
    for length in lengths:
        file="E:/LAND USE/data/" + city + "/" + city + "_" + str(length) + "cell_point_tree#" + str(min_count) + "_new.csv"
        data=pd.read_csv(file)
        cell_num=len(data)
        DF_map=get_df_map(data) 
        cell_category_num={i:data.loc[i][1] for i in range(len(data))} 
        all_data=[]
        for i in tqdm(range(len(data))):
            col=[data.loc[i,'0']]
            for j in range(2,len(data.loc[0])):
                TF=data.loc[i,str(j)]/data.loc[i,'1']
#                 IDF=log(1+cell_num/(DF_map[j]))
                IDF=log(cell_num/(DF_map[j]+1))
                col.append(TF*IDF)
            all_data.append(col)
        pd_data=pd.DataFrame(all_data)
        file1="E:/LAND USE/data/" + city + "/" + city + "_" + str(length) + "cell_poi_tf-idf#" + str(min_count) + "_new.csv"
        pd_data.to_csv(file1,index=None)
```
